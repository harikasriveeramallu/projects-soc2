LINEAR REGRESSION

```python
from sklearn.linear_model import LinearRegression
LR=LinearRegression()
import numpy as np
import pandas as pd
```

```python
mydf=pd.read_csv("/content/apple_quality.csv")
mydf
```

```python
mydf.dropna(inplace=True)
```

```python
s=np.array(mydf["Size"]).reshape(-1,1)
c=mydf["Sweetness"]
LR.fit(s,c)
```

```python
LR.predict([[1.9]])
```

LOGISTIC REGRESSION

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression()
```

DECISION TREE

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
A_col=["Weight","Sweetness","Crunchiness","Juiciness"]
A=mydf[A_col]
Target=mydf.Quality
```

```python
model=DecisionTreeClassifier()
model.fit(A,Target)
```

```python
a=float(input("Enter Weight of the Apple"))
b=float(input("Enter Sweetness of the Apple"))
c=float(input("Enter Crunchiness of the Apple"))
d=float(input("Enter Juiciness of the Apple"))
p=model.predict([[a,b,c,d]])
if p=='good':
  print("Good Quality")
else:
  print("Bad Quality")
```

```python
from sklearn.tree import export_graphviz,plot_tree
```

```python
plot_tree(model)
```

RANDOM FOREST

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
A_col=["Weight","Sweetness","Crunchiness","Juiciness"]
A=mydf[A_col]
Target=mydf.Quality
```

```python
rf=RandomForestClassifier(n_estimators=15)
rf.fit(A,Target)
```

```python
a=float(input("Enter Weight of the Apple"))
b=float(input("Enter Sweetness of the Apple"))
c=float(input("Enter Crunchiness of the Apple"))
d=float(input("Enter Juiciness of the Apple"))
p=model.predict([[a,b,c,d]])
if p=='good':
  print("Good Quality")
else:
  print("Bad Quality")
```

```python
import os
output_dir="tree_visualizations"
os.makedirs(output_dir,exist_ok=True)
```

```python
rf.estimators_
```

```python
for i,tree in enumerate(rf.estimators_):
  tree_dot_file=os.path.join(output_dir,f"tree_{i}.dot")
  tree_png_file=os.path.join(output_dir,f"tree_{i}.png")
```

```python
from sklearn.tree import export_graphviz
export_graphviz(tree, out_file=tree_dot_file, feature_names=["Weight","Sweetness","Crunchiness","Juiciness"],
                    class_names=[str(cls) for cls in rf.classes_], filled=True, rounded=True)
```

```python
command=f"dot -Tpng {tree_dot_file} -o {tree_png_file}"
os.system(command)
print(f"Tree{i} visualization saved to {tree_png_file}")
```
